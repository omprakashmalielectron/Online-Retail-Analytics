```python
# Q1: Retrieve top 5 highest value completed orders (order_id, customer_id, order_amount, order_date)
q1="""
    SELECT 
        o.order_id, 
        o.customer_id,
        SUM(oi.quantity*oi.unit_price) as order_amount,
        o.order_date
    FROM orders as o
    LEFT JOIN order_items as oi ON oi.order_id=o.order_id
    Where o.status='Completed'
    GROUP BY  o.order_id,o.customer_id,o.order_date
    ORDER BY order_amount DESC
    LIMIT 5
"""
pd.read_sql(q1,engine)
```

```python
# Q2: List customers with more than 2 orders (customer_id, full_name, total_orders)
q2="""
    SELECT 
        c.customer_id, 
        c.full_name,
        COUNT(o.order_id) as total_orders
    FROM customers as c
    LEFT JOIN orders as o 
        ON o.customer_id=c.customer_id
    GROUP BY c.customer_id,c.full_name
    HAVING COUNT(o.order_id)>2
"""
pd.read_sql(q2,engine)
```

```python
#Q3: Get customers who never placed an order using LEFT JOIN (customer_id, full_name, signup_date)
q3="""
    SELECT 
        c.customer_id,
        c.full_name,
        c.signup_date
    FROM customers as c
    LEFT JOIN orders as o ON c.customer_id=o.customer_id
    WHERE o.order_id is NULL
"""
pd.read_sql(q3,engine)
```

```python
# Q4: Show most expensive product purchased by each customer (customer_id, product_name, price)
# giving most expensive orders which are actually completed
q4="""
    SELECT 
        c.customer_id,
        p.product_name,
        p.price as max_price
    FROM customers as c
    INNER JOIN orders as o on c.customer_id=o.customer_id
    INNER JOIN order_items as oi on oi.order_id=o.order_id
    INNER JOIN products as p on p.product_id=oi.product_id
    WHERE p.price=(
        SELECT Max(p1.price)
        FROM orders as o1
        INNER JOIN order_items as oi1 on oi1.order_id=o1.order_id
        INNER JOIN products as p1 on p1.product_id=oi1.product_id
        WHERE c.customer_id=o1.customer_id
    ) AND o.status='Completed'
    ORDER BY customer_id
"""
pd.read_sql(q4,engine)
```

```python
# Q5: Find order with largest number of items (order_id, total_items)
q5="""
    SELECT
        o.order_id,
        COUNT(oi.order_item_id) as total_items
    FROM orders as o
    INNER JOIN order_items as oi on oi.order_id=o.order_id
    GROUP BY o.order_id
    ORDER BY COUNT(oi.order_item_id) DESC
    LIMIT 1
"""
pd.read_sql(q5,engine)
```

```python
# Q6: List customers who ordered within 30 days of signup (customer_id, full_name, signup_date, first_order_date, 
# days_difference)
# data is very inconsistent as there are orders that are ordered before signup date
q6="""
    SELECT
        c.customer_id,
        c.full_name,
        c.signup_date,
        o.order_date as first_order_date,
        (o.order_date-c.signup_date) as days_difference
    FROM customers as c
    INNER JOIN orders as o ON c.customer_id=o.customer_id
    WHERE o.order_date=(
        SELECT
            Min(o1.order_date)
            FROM orders as o1
            WHERE o1.customer_id=c.customer_id
    )
    AND (o.order_date-c.signup_date)<=30
"""
pd.read_sql(q6,engine)
```

```python
# Q7: Count orders per status per country (country, status, order_count)
q7="""
    SELECT
        c.country,
        o.status,
        COUNT(o.order_id) as order_count
    FROM orders as o
    INNER JOIN customers as c ON c.customer_id=o.customer_id
    GROUP BY c.country,o.status
    ORDER BY c.country
"""
pd.read_sql(q7,engine)
```

```python
#Q8: Calculate Customer Lifetime Value (customer_id, full_name, total_revenue)
q8="""
    SELECT
        c.customer_id,
        c.full_name,
        COALESCE(SUM(oi.quantity*oi.unit_price)) as total_revenue
    FROM customers as c 
    LEFT JOIN orders as o on o.customer_id=c.customer_id
    LEFT JOIN order_items as oi on oi.order_id=o.order_id
    GROUP BY c.customer_id,c.full_name
    order by c.customer_id
"""
pd.read_sql(q8,engine)
```

```python
# Q9: Identify repeat customers - more than 1 order per month (customer_id, full_name, order_month, monthly_order_count)
q9="""
    SELECT 
        c.customer_id,
        c.full_name,
        DATE_TRUNC('month',o.order_date) as order_month,
        COUNT(o.order_id) as monthly_order_count
    FROM customers as c
    LEFT JOIN orders as o on c.customer_id=o.customer_id
    GROUP BY c.customer_id,c.full_name,DATE_TRUNC('month',o.order_date)
    HAVING COUNT(o.order_id)>1
"""
pd.read_sql(q9,engine)
```

```python
# Q10: Monthly revenue trend (year, month, total_revenue)
q10="""
    SELECT
        DATE_TRUNC('year',o.order_date) as year,
        DATE_TRUNC('month',o.order_date) as month,
        SUM(oi.quantity*oi.unit_price) as total_revenue
    FROM customers as c
    INNER JOIN orders as o on c.customer_id=o.customer_id
    INNER JOIN order_items as oi on oi.order_id=o.order_id
    GROUP BY DATE_TRUNC('month',o.order_date),DATE_TRUNC('year',o.order_date)
"""
pd.read_sql(q10,engine)
```

```python
# Q11: Most returned products where status=Returned (product_id, product_name, return_count)
q11="""
    SELECT
        p.product_id,
        p.product_name,
        COUNT(o.order_id) as return_count
    FROM products as p
    INNER JOIN order_items as oi on oi.product_id=p.product_id
    INNER JOIN orders as o on o.order_id=oi.order_id
    WHERE o.status='Returned'
    GROUP BY p.product_id,p.product_name
"""
pd.read_sql(q11,engine)
```

```python
# Q12: Top 3 revenue-generating categories (category, total_revenue)
q12="""
    SELECT
        p.category,
        SUM(oi.quantity*oi.unit_price) as total_revenue
    FROM products as p
    INNER JOIN order_items as oi ON oi.product_id=p.product_id
    GROUP BY p.category
    ORDER BY total_revenue DESC
    LIMIT 3
"""
pd.read_sql(q12,engine)
```

```python
# Q13: Products frequently bought together - basket analysis (product_id_1, product_id_2, times_bought_together)
q13="""
    SELECT
        oi.product_id as product_id_1,
        oi1.product_id as product_id_2,
        COUNT(oi.order_id) as times_bought_together
    FROM order_items as oi
    INNER JOIN order_items as oi1 on oi.order_id=oi1.order_id 
            AND oi.product_id<oi1.product_id
    GROUP BY oi.product_id,oi1.product_id
"""
pd.read_sql(q13,engine)
```

```python
# Q15: Net Revenue After Discounts & Tax monthly country-wise (country, year, month, gross_revenue, total_discount, 
#      total_tax, net_revenue)
q15="""
    WITH monthly_revenue as(
        SELECT
            c.country,
            DATE_TRUNC('year',o.order_date) as year,
            DATE_TRUNC('month',o.order_date) as month,
            SUM(o.discount_amount) as total_discount,
            SUM(o.tax_amount) as total_tax,
            SUM(oi.quantity*oi.unit_price) as total_amt
        FROM customers as c
        INNER JOIN orders as o on o.customer_id=c.customer_id
        INNER JOIN order_items as oi on oi.order_id=o.order_id
        GROUP BY c.country,DATE_TRUNC('year',o.order_date),DATE_TRUNC('month',o.order_date)
    )
    SELECT
        m.country,
        m.year,
        m.month,
        m.total_discount,
        m.total_tax,
        m.total_amt-m.total_discount-m.total_tax as net_revenue
    FROM monthly_revenue as m
"""
pd.read_sql(q15,engine)
```

```python
# Q16: Discount Sensitivity - customers who only order with discounts (customer_id, full_name, total_orders, 
#      discounted_orders)
### IN this query i essentially took first quartile as a threshold to classify an order as a discounted order or not
### and the result i got 750 rows which is 75% of all the orders(1000 rows) which is expected give that each customer
### have placed only single order
q16="""
    WITH discount_threshold AS(
        SELECT
            PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY o.discount_amount) as q1_discount
        FROM orders as o
    )
    SELECT
        c.customer_id,
        c.full_name,
        COUNT(o.order_id) as total_orders,
        COUNT(
            CASE 
                WHEN o.discount_amount>dt.q1_discount THEN 1
            END
        )
        FROM customers as c
        INNER JOIN orders as o on c.customer_id=o.customer_id
        CROSS JOIN discount_threshold as dt
        GROUP BY c.customer_id,c.full_name
        HAVING COUNT(o.order_id)=COUNT(
                 CASE 
                     WHEN o.discount_amount>dt.q1_discount THEN 1
                 END
        )
"""
pd.read_sql(q16,engine)
```

```python
# Q19: Brand Affinity per Customer (customer_id, brand, brand_revenue)
q19="""
    SELECT
        c.customer_id,
        p.brand,
        SUM(oi.quantity*oi.unit_price) as brand_revenue
    FROM customers as c
    INNER JOIN orders as o on o.customer_id=c.customer_id
    INNER JOIN order_items as oi on oi.order_id=o.order_id
    INNER JOIN products as p on p.product_id=oi.product_id
    GROUP BY c.customer_id,p.brand
    ORDER BY brand_revenue
"""
pd.read_sql(q19,engine)
```
